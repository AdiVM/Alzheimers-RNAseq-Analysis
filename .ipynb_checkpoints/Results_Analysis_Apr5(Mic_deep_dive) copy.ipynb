```python
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Subset just Microglia cells (if needed)
mic_metadata = metadata[metadata['broad.cell.type'] == 'Mic']
mic_matrix = gene_matrix.loc[mic_metadata['TAG']]

# Step 2: Join metadata and expression
mic_metadata = mic_metadata.set_index('TAG')
mic_data = mic_matrix.copy()
mic_data['alzheimers_or_control'] = mic_metadata['alzheimers_or_control']

# Step 3: Extract expression values for C1QA and SPP1
ad_cells_spp1 = mic_data[mic_data['alzheimers_or_control'] == 1]['SPP1']
ctrl_cells_spp1 = mic_data[mic_data['alzheimers_or_control'] == 0]['SPP1']

ad_cells_c1qa = mic_data[mic_data['alzheimers_or_control'] == 1]['C1QA']
ctrl_cells_c1qa = mic_data[mic_data['alzheimers_or_control'] == 0]['C1QA']

# Step 4: Mann-Whitney U Test
_, pval_spp1 = mannwhitneyu(ad_cells_spp1, ctrl_cells_spp1, alternative='two-sided')
_, pval_c1qa = mannwhitneyu(ad_cells_c1qa, ctrl_cells_c1qa, alternative='two-sided')

print(f"SPP1: p = {pval_spp1:.3e}")
print(f"C1QA: p = {pval_c1qa:.3e}")

# Step 5: Optional – Plot expression levels
sns.stripplot(data=mic_data, x='alzheimers_or_control', y='SPP1', alpha=0.4)
plt.title(f"SPP1 Expression (p = {pval_spp1:.3e})")
plt.xlabel("Alzheimer's (1) vs Control (0)")
plt.show()

sns.stripplot(data=mic_data, x='alzheimers_or_control', y='C1QA', alpha=0.4)
plt.title(f"C1QA Expression (p = {pval_c1qa:.3e})")
plt.xlabel("Alzheimer's (1) vs Control (0)")
plt.show()
```

Goal: I'm trying to see if gene expression of C1QA and SPP1 is different across AD and control cells.

```python
# First check if there are genes diffefernetially expressed between AD and control cells
from scipy.stats import mannwhitneyu
import pandas as pd

# Compute which genes are expressed in >10% of cells
min_cells = int(0.10 * len(mic_matrix))
expressed_genes = mic_matrix.columns[(mic_matrix > 0).sum(axis=0) >= min_cells]

# Group cells
ad_cells = mic_metadata[mic_metadata['alzheimers_or_control'] == 1].index
ctrl_cells = mic_metadata[mic_metadata['alzheimers_or_control'] == 0].index

# Store results
deg_results = []

for gene in expressed_genes:
    ad_expr = mic_matrix.loc[ad_cells, gene]
    ctrl_expr = mic_matrix.loc[ctrl_cells, gene]
    
    if ad_expr.sum() == 0 and ctrl_expr.sum() == 0:
        continue

    stat, p = mannwhitneyu(ad_expr, ctrl_expr, alternative='two-sided')
    effect = ad_expr.median() - ctrl_expr.median()
    deg_results.append({'gene': gene, 'p_value': p, 'effect_size': effect})

deg_df = pd.DataFrame(deg_results).sort_values('p_value')
```

```python
# FDR correction
deg_df['fdr'] = pd.Series(deg_df['p_value']).rank(method='min') / len(deg_df) * 0.05

# Filter for statistically significant genes
significant_degs = deg_df[deg_df['p_value'] < 1e-3]

# Print the top 10 by p-value
print("Top 10 DEGs by p-value:")
print(deg_df.head(10))
```

```python
mic_metadata = metadata[metadata['Subcluster'].isin(['Mic0', 'Mic1'])].copy()
mic_matrix = gene_matrix.loc[mic_metadata['TAG']]
mic_metadata = mic_metadata.set_index('TAG')
mic_data = mic_matrix.copy()
mic_data['Subcluster'] = mic_metadata['Subcluster']
```

```python
# Now check if these genes are differentially expressed between subclusters

from scipy.stats import mannwhitneyu

top_ad_genes = ['PLXDC2', 'SRGN', 'OLR1', 'SLC1A3', 'SDK1', 
                'RTTN', 'P2RY12', 'HS3ST4', 'SOCS6', 'KCNMA1']

results = []
for gene in top_ad_genes:
    mic0_expr = mic_data[mic_data['Subcluster'] == 'Mic0'][gene]
    mic1_expr = mic_data[mic_data['Subcluster'] == 'Mic1'][gene]

    # Skip genes with all zeros
    if mic0_expr.sum() == 0 and mic1_expr.sum() == 0:
        continue

    stat, pval = mannwhitneyu(mic0_expr, mic1_expr, alternative='two-sided')
    effect = mic1_expr.median() - mic0_expr.median()

    results.append({'gene': gene, 'pval': pval, 'effect_size': effect})

diff_subcluster_df = pd.DataFrame(results).sort_values('pval')
```

```python
import scanpy as sc
import anndata

# Create AnnData from mic_matrix
adata = sc.AnnData(mic_matrix)
mic_metadata['TAG'] = mic_metadata.index

# Add metadata
adata.obs = mic_metadata.set_index('TAG').loc[adata.obs_names]
```

```python
import scanpy as sc

# 1. Use existing subclusters as group labels for PAGA
adata.obs['subcluster'] = adata.obs['Subcluster']

# 2. Run PCA (if not already done)
sc.tl.pca(adata)

# 3. Compute neighborhood graph
sc.pp.neighbors(adata, n_neighbors=15, n_pcs=30)

# 4. Compute UMAP (for visualization) and run PAGA
sc.tl.umap(adata)
sc.tl.paga(adata, groups='subcluster')  # Use your Mic0, Mic1 labels

# 5. Visualize PAGA graph over UMAP
sc.pl.paga(adata, color=['subcluster'], threshold=0.05)
sc.pl.umap(adata, color='subcluster', legend_loc='on data')
sc.pl.paga_compare(adata, basis='umap', color='subcluster')

# 6. Choose a Mic cluster as root (e.g. Mic0 as starting point)
# 6. Choose a Mic cluster as root (e.g. Mic0 as starting point)
root_cell = adata.obs.query("subcluster == 'Mic0'").index[0]  # pick any cell from Mic0
adata.uns['iroot'] = np.where(adata.obs_names == root_cell)[0][0]  # this sets the root index

# 7. Run diffusion pseudotime (DPT)
sc.tl.dpt(adata)

# 8. Visualize pseudotime
# 8. Visualize pseudotime
sc.pl.umap(adata, color='dpt_pseudotime', cmap='viridis')
```
