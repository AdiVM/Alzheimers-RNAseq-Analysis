```python
from gprofiler import GProfiler

# Your gene list
genes_of_interest =  genes_in_module_6

# Run g:Profiler
gp = GProfiler(return_dataframe=True)
result = gp.profile(
    organism='hsapiens',
    query=genes_of_interest,
    sources=['GO:BP', 'GO:MF', 'GO:CC', 'KEGG', 'REAC']
)
print(result.columns)
```

```python
# Show top 10 enriched pathways
result[['source', 'name', 'p_value', 'intersection_size']].head(10)

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Clean + rank results
top_pathways = result.sort_values('p_value').head(12).copy()
top_pathways['-log10(p_value)'] = -np.log10(top_pathways['p_value'])

# Plot like your target example
plt.figure(figsize=(6, 5))
sns.barplot(
    data=top_pathways,
    x='-log10(p_value)',
    y='name',
    color='steelblue'
)

# Style
plt.xlabel(r'$-\log_{10}(\it{P}\text{-value})$', fontsize=12)
plt.ylabel('')
plt.title('Top Enriched Pathways', fontsize=14)
plt.gca().invert_yaxis()
sns.despine()
plt.tight_layout()
plt.show()
```

```python
import joblib
import os
import matplotlib.pyplot as plt
import numpy as np

# Map model names to directories
model_dirs = {
    'Combined': '/n/groups/patel/adithya/Alz_Outputs/Final_Outputs/Cell_on_cell_nopmi_rfe/Mic/',
    'GeneOnly': '/n/groups/patel/adithya/Alz_Outputs/Final_Outputs/Cell_on_cell_genes_rfe/Mic/',
    'DemoOnly': '/n/groups/patel/adithya/Alz_Outputs/Final_Outputs/Cell_on_cell_demo_nopmi/Mic/',
}
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Helper function
def plot_roc_curve(true_labels, predicted_probs, label, color):
    fpr, tpr, _ = roc_curve(true_labels, predicted_probs)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{label} (AUC = {roc_auc:.2f})", color=color)
    return roc_auc

# Plot
plt.figure(figsize=(8, 6))

auc_cb = plot_roc_curve(Mic_cb_pred['true_label'], Mic_cb_pred['predicted_proba'], 'Combined Model', 'blue')
auc_cg = plot_roc_curve(Mic_cg_pred['true_label'], Mic_cg_pred['predicted_proba'], 'Gene Model', 'green')
auc_cd = plot_roc_curve(Mic_cd_pred['true_label'], Mic_cd_pred['predicted_proba'], 'Demographics Model', 'orange')

plt.plot([0, 1], [0, 1], 'k--', label='Random')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Mic Models')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def extract_nonzero_feature_importances(model_path):
    model = joblib.load(model_path)
    
    # Get feature names and importance values
    feature_names = np.array(model.feature_names_in_)
    importance_values = model.feature_importances_

    # Handle nested array from some FLAML versions
    if hasattr(importance_values, "__len__") and len(importance_values) == 1:
        importance_values = importance_values[0]

    # Filter and return non-zero features
    nonzero_mask = importance_values != 0
    return feature_names[nonzero_mask], importance_values[nonzero_mask]

# Collect and print non-zero importance features for each model
for model_name, model_path in model_dirs.items():
    clf_path = os.path.join(model_path, "maximal_classifier.joblib")
    
    features, importances = extract_nonzero_feature_importances(clf_path)
    
    # Sort by importance
    sorted_idx = np.argsort(importances)[::-1]
    top_features = features[sorted_idx]
    top_importances = importances[sorted_idx]

    # Create DataFrame
    df = pd.DataFrame({
        'feature': top_features,
        'importance': top_importances
    })

    print(f"Top importance features for {model_name} model:")
    print(df.to_string(index=False))
```

```python
import joblib
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Map model names to directories
model_dirs = {
    'Combined': '/n/groups/patel/adithya/Alz_Outputs/Final_Outputs/Cell_on_cell_nopmi_rfe/Ast/',
    'GeneOnly': '/n/groups/patel/adithya/Alz_Outputs/Final_Outputs/Cell_on_cell_genes_rfe/Ast/',
    'DemoOnly': '/n/groups/patel/adithya/Alz_Outputs/Final_Outputs/Cell_on_cell_demo_nopmi/Ast/',
}

# Set importance threshold for GeneOnly model
gene_only_threshold = 4

def extract_nonzero_feature_importances(model_path):
    model = joblib.load(model_path)
    
    # Get feature names and importance values
    feature_names = np.array(model.feature_names_in_)
    importance_values = model.feature_importances_

    # Handle nested array from some FLAML versions
    if hasattr(importance_values, "__len__") and len(importance_values) == 1:
        importance_values = importance_values[0]

    return feature_names, importance_values

# Plot and print
for model_name, model_path in model_dirs.items():
    clf_path = os.path.join(model_path, "maximal_classifier.joblib")
    
    features, importances = extract_nonzero_feature_importances(clf_path)

    # Apply filtering
    if model_name == 'GeneOnly':
        mask = importances > gene_only_threshold
    else:
        mask = importances != 0

    features = features[mask]
    importances = importances[mask]

    if len(features) == 0:
        print(f"No features to plot for {model_name} model.")
        continue

    # Sort by importance
    sorted_idx = np.argsort(importances)[::-1]
    top_features = features[sorted_idx]
    top_importances = importances[sorted_idx]

    # Create DataFrame
    df = pd.DataFrame({
        'feature': top_features,
        'importance': top_importances
    })

    print(f"\nTop importance features for {model_name} model:")
    print(df.to_string(index=False))

    # Plot
    plt.figure(figsize=(10, 5))
    plt.barh(df['feature'][::-1], df['importance'][::-1])  # reverse for descending top-down
    plt.title(f"Feature Importances - {model_name} Model")
    plt.xlabel("Importance")
    plt.tight_layout()
    plt.grid(axis='x')
    plt.show()
```

```python
import os
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

def roc_imp_feature(cell_type, gene_only_threshold=2):
    # === Base model paths ===
    base_paths = {
        'Combined': '/n/groups/patel/adithya/Alz_Outputs/Final_Outputs/Cell_on_cell_nopmi_rfe/',
        'GeneOnly': '/n/groups/patel/adithya/Alz_Outputs/Final_Outputs/Cell_on_cell_genes_rfe/',
        'DemoOnly': '/n/groups/patel/adithya/Alz_Outputs/Final_Outputs/Cell_on_cell_demo_nopmi/',
    }
    model_dirs = {key: os.path.join(path, cell_type) for key, path in base_paths.items()}

    # === Load prediction data ===
    pred_files = {
        name: os.path.join(model_dirs[name], 'test_predictions.csv')
        for name in model_dirs
    }
    pred_data = {key: pd.read_csv(path) for key, path in pred_files.items()}

    # === Plot ROC curves ===
    def plot_roc_curve(true_labels, predicted_probs, label, color):
        fpr, tpr, _ = roc_curve(true_labels, predicted_probs)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f"{label} (AUC = {roc_auc:.2f})", color=color)
        return roc_auc

    plt.figure(figsize=(8, 6))
    colors = {'Combined': 'blue', 'GeneOnly': 'green', 'DemoOnly': 'orange'}
    for model_name in ['Combined', 'GeneOnly', 'DemoOnly']:
        df = pred_data[model_name]
        plot_roc_curve(df['true_label'], df['predicted_proba'], f'{model_name} Model', colors[model_name])

    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curves for {cell_type} Models')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # === Extract and plot feature importances ===
    def extract_nonzero_feature_importances(model_path):
        model = joblib.load(model_path)
        feature_names = np.array(model.feature_names_in_)
        importance_values = model.feature_importances_

        if hasattr(importance_values, "__len__") and len(importance_values) == 1:
            importance_values = importance_values[0]

        return feature_names, importance_values

    for model_name, model_path in model_dirs.items():
        clf_path = os.path.join(model_path, "maximal_classifier.joblib")
        features, importances = extract_nonzero_feature_importances(clf_path)

        # Apply filtering
        if model_name == 'GeneOnly':
            mask = importances > gene_only_threshold
        else:
            mask = importances != 0

        features = features[mask]
        importances = importances[mask]

        if len(features) == 0:
            print(f"No features to plot for {model_name} model.")
            continue

        sorted_idx = np.argsort(importances)[::-1]
        top_features = features[sorted_idx]
        top_importances = importances[sorted_idx]

        df = pd.DataFrame({'feature': top_features, 'importance': top_importances})
        print(f"\nTop features for {model_name} model:\n{df.to_string(index=False)}")

        plt.figure(figsize=(10, 5))
        plt.barh(df['feature'][::-1], df['importance'][::-1])
        plt.title(f"Feature Importances - {model_name} Model")
        plt.xlabel("Importance")
        plt.tight_layout()
        plt.grid(axis='x')
        plt.show()
roc_imp_feature("In")
```
