```python
# Now check if these genes are differentially expressed between subclusters
from scipy.stats import mannwhitneyu

top_ad_genes = ['PLXDC2', 'SRGN', 'OLR1', 'SLC1A3', 'SDK1', 
                'RTTN', 'P2RY12', 'HS3ST4', 'SOCS6', 'KCNMA1']

results = []
for gene in top_ad_genes:
    mic0_expr = mic_data[mic_data['Subcluster'] == 'Mic0'][gene]
    mic1_expr = mic_data[mic_data['Subcluster'] == 'Mic1'][gene]

    # Skip genes with all zeros
    if mic0_expr.sum() == 0 and mic1_expr.sum() == 0:
        continue

    stat, pval = mannwhitneyu(mic0_expr, mic1_expr, alternative='two-sided')
    effect = mic1_expr.median() - mic0_expr.median()

    results.append({'gene': gene, 'pval': pval, 'effect_size': effect})

diff_subcluster_df = pd.DataFrame(results).sort_values('pval')
```

```python
print(diff_subcluster_df)
```

We performed targetted differential expression analysis between Mic0 and Mic1 subclusters and found that the above genes expression was signficantly different between them.

```python
import matplotlib.pyplot as plt
import numpy as np

# Use your existing dataframe
deg_df = df.copy()

# Calculate -log10(p-value)
deg_df['-log10(pval)'] = -np.log10(deg_df['pval'] + 1e-300)  # avoid log(0)

# Volcano plot
plt.figure(figsize=(8, 6))
plt.scatter(deg_df['log2fc'], deg_df['-log10(pval)'], 
            c='grey', s=10, alpha=0.7)

# Genes of intrest
genes_of_interest = ['PLXDC2', 'SLC1A3', 'SDK1', 'KCNMA1', 'CSGALNACT1', 'OLR1', 'P2RY12', 'HS3ST4', 'DOCK10', 'PICALM']
for gene in genes_of_interest:
    if gene in deg_df['gene'].values:
        row = deg_df[deg_df['gene'] == gene].iloc[0]
        plt.scatter(row['log2fc'], row['-log10(pval)'], color='red', s=40)
        plt.text(row['log2fc'], row['-log10(pval)'] + 0.1, gene, fontsize=9)

# Annotate thresholds
plt.axhline(-np.log10(0.05), color='black', linestyle='--', linewidth=1)
plt.axvline(1, color='black', linestyle='--', linewidth=1)
plt.axvline(-1, color='black', linestyle='--', linewidth=1)

# Labels
plt.xlabel('Log2 Fold Change')
plt.ylabel('-log10(p-value)')
plt.title('Volcano Plot (AD vs Control)')
plt.tight_layout()
plt.show()
```
