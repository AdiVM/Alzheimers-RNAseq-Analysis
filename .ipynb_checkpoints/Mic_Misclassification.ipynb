```python
pred_df = prediction_file
meta_df = metadata_file
```

```python
# Filter for false negatives and false positives
misclassified_df = pred_df[
    ((pred_df['true_label'] == 1) & (pred_df['predicted_label'] == 0)) |  # FN
    ((pred_df['true_label'] == 0) & (pred_df['predicted_label'] == 1))    # FP
]
```

```python
misclassified_merged_df = misclassified_df.merge(meta_df[['TAG', 'Subcluster']], on='TAG', how='left')
```

```python
# This looks right
import matplotlib.pyplot as plt

# Count misclassified cells per cluster
cluster_counts = misclassified_merged_df['Subcluster'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 6))
cluster_counts.plot(kind='bar')
plt.xlabel('Subcluster')
plt.ylabel('Number of Misclassified Cells')
plt.title('Misclassified Cells per Subcluster')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Filter for just microglial cells
mic_meta_df = meta_df[meta_df['broad.cell.type'] == 'Mic']

# Calculate total cells per cluster (among microglia only)
total_cells_per_cluster = mic_meta_df['Subcluster'].value_counts().sort_index()
print("Total number of Mic cells per cluster:\n", total_cells_per_cluster)

print("The above values included both training and testing Mic cells while this only looks at testing Mic cells")

# Misclassified cells per cluster (already filtered by prediction frame, assuming it's all Mic)
misclassified_per_cluster = misclassified_merged_df['Subcluster'].value_counts().sort_index()
print("Total number of misclassified Mic cells per cluster:\n", misclassified_per_cluster)
```

```python
# Total number of predictions
total_cells = len(pred_df)
print(f"Total number of cells: {total_cells}")

# Number of misclassified predictions
num_misclassified = (pred_df['true_label'] != pred_df['predicted_label']).sum()

# Percentage of misclassified cells
percent_misclassified = (num_misclassified / total_cells) * 100

print(f"Total misclassified cells: {num_misclassified}")
print(f"Percentage of misclassified cells: {percent_misclassified:.2f}%")
```

```python
# Get all test cells (misclassified or not) from prediction dataframe merged with metadata
test_merged_df = pred_df.merge(meta_df[['TAG', 'Subcluster']], on='TAG')

# Get the total number of test cells in each subcluster
total_test_cells_per_cluster = test_merged_df['Subcluster'].value_counts().sort_index()

# Get number of misclassified cells per subcluster
misclassified_per_cluster = misclassified_merged_df['Subcluster'].value_counts().sort_index()

# Align the index and calculate percentage
misclassified_percentage = (misclassified_per_cluster / total_test_cells_per_cluster) * 100

# Plot
plt.figure(figsize=(10, 6))
misclassified_percentage.plot(kind='bar')
plt.xlabel('Subcluster')
plt.ylabel('Percentage of Misclassified Cells')
plt.title('Misclassification Rate per Subcluster (Test Set)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Confident correct cells
confident_correct_ad = pred_df[(pred_df['true_label'] == 1) & (pred_df['predicted_proba'] > 0.5)]
confident_correct_ctrl = pred_df[(pred_df['true_label'] == 0) & (pred_df['predicted_proba'] <= 0.2)]
confident_correct = pd.concat([confident_correct_ad, confident_correct_ctrl])

# Confident incorrect cells
confident_incorrect_ad = pred_df[(pred_df['true_label'] == 1) & (pred_df['predicted_proba'] <= 0.2)]
confident_incorrect_ctrl = pred_df[(pred_df['true_label'] == 0) & (pred_df['predicted_proba'] >= 0.7)]
confident_incorrect = pd.concat([confident_incorrect_ad, confident_incorrect_ctrl])

# Bring in Subcluster info
confident_correct = confident_correct.merge(meta_df[['TAG', 'Subcluster']], on='TAG', how='left')
confident_incorrect = confident_incorrect.merge(meta_df[['TAG', 'Subcluster']], on='TAG', how='left')

# Count cluster distribution
correct_counts = confident_correct['Subcluster'].value_counts().sort_index()
incorrect_counts = confident_incorrect['Subcluster'].value_counts().sort_index()

# Plot side-by-side bar chart
import matplotlib.pyplot as plt

clusters = sorted(set(correct_counts.index).union(incorrect_counts.index))
correct_vals = [correct_counts.get(c, 0) for c in clusters]
incorrect_vals = [incorrect_counts.get(c, 0) for c in clusters]

x = range(len(clusters))
plt.figure(figsize=(10,6))
plt.bar(x, correct_vals, width=0.4, label='Confident Correct', align='center')
plt.bar([i + 0.4 for i in x], incorrect_vals, width=0.4, label='Confident Incorrect', align='center')
plt.xticks([i + 0.2 for i in x], clusters, rotation=45)
plt.ylabel("Cell Count")
plt.title("Subcluster Distribution of Confident Predictions")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Split into four groups
confident_correct_ad = pred_df[(pred_df['true_label'] == 1) & (pred_df['predicted_proba'] >= 0.5)]
confident_correct_ctrl = pred_df[(pred_df['true_label'] == 0) & (pred_df['predicted_proba'] <= 0.35)]
confident_incorrect_ad = pred_df[(pred_df['true_label'] == 1) & (pred_df['predicted_proba'] <= 0.2)]
confident_incorrect_ctrl = pred_df[(pred_df['true_label'] == 0) & (pred_df['predicted_proba'] >= 0.5)]

# Add subcluster info
confident_correct_ad = confident_correct_ad.merge(meta_df[['TAG', 'Subcluster']], on='TAG', how='left')
confident_correct_ctrl = confident_correct_ctrl.merge(meta_df[['TAG', 'Subcluster']], on='TAG', how='left')
confident_incorrect_ad = confident_incorrect_ad.merge(meta_df[['TAG', 'Subcluster']], on='TAG', how='left')
confident_incorrect_ctrl = confident_incorrect_ctrl.merge(meta_df[['TAG', 'Subcluster']], on='TAG', how='left')

# Count per cluster
cc_ad_counts = confident_correct_ad['Subcluster'].value_counts().sort_index()
cc_ctrl_counts = confident_correct_ctrl['Subcluster'].value_counts().sort_index()
ci_ad_counts = confident_incorrect_ad['Subcluster'].value_counts().sort_index()
ci_ctrl_counts = confident_incorrect_ctrl['Subcluster'].value_counts().sort_index()

# Get full set of clusters
clusters = sorted(set(
    cc_ad_counts.index
).union(cc_ctrl_counts.index).union(ci_ad_counts.index).union(ci_ctrl_counts.index))

# Get values for each bar
cc_ad_vals = [cc_ad_counts.get(c, 0) for c in clusters]
cc_ctrl_vals = [cc_ctrl_counts.get(c, 0) for c in clusters]
ci_ad_vals = [ci_ad_counts.get(c, 0) for c in clusters]
ci_ctrl_vals = [ci_ctrl_counts.get(c, 0) for c in clusters]

# Plot
import matplotlib.pyplot as plt

x = range(len(clusters))
bar_width = 0.2

plt.figure(figsize=(12, 6))
plt.bar([i - 1.5 * bar_width for i in x], cc_ad_vals, width=bar_width, label='Correct Case')
plt.bar([i - 0.5 * bar_width for i in x], cc_ctrl_vals, width=bar_width, label='Correct Control')
plt.bar([i + 0.5 * bar_width for i in x], ci_ad_vals, width=bar_width, label='Incorrect Case')
plt.bar([i + 1.5 * bar_width for i in x], ci_ctrl_vals, width=bar_width, label='Incorrect Control')

plt.xticks(x, clusters, rotation=45)
plt.ylabel("Cell Count")
plt.title("Subcluster Distribution of Confident Predictions (4 Categories)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
merged_pred = pred_df.merge(meta_df[['TAG', 'Subcluster']], on='TAG', how='left')
from sklearn.metrics import roc_auc_score
# Group by Subcluster
cluster_aucs = {}

for cluster, group in merged_pred.groupby('Subcluster'):
    auc = roc_auc_score(group['true_label'], group['predicted_proba'])
    cluster_aucs[cluster] = auc

# Print AUC
for cluster, auc in cluster_aucs.items():
    print(f"{cluster}: AUC = {auc:.3f}")
```
