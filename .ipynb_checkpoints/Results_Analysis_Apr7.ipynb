```python
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Subset just Microglia cells (if needed)
mic_metadata = metadata[metadata['broad.cell.type'] == 'Mic']
mic_matrix = gene_matrix.loc[mic_metadata['TAG']]

# Step 2: Join metadata and expression
mic_metadata = mic_metadata.set_index('TAG')
mic_data = mic_matrix.copy()
mic_data['alzheimers_or_control'] = mic_metadata['alzheimers_or_control']
```

```python
# First check if there are genes diffefernetially expressed between AD and control cells
from scipy.stats import mannwhitneyu
import pandas as pd

# Compute which genes are expressed in >10% of cells
min_cells = int(0.10 * len(mic_matrix))
expressed_genes = mic_matrix.columns[(mic_matrix > 0).sum(axis=0) >= min_cells]

# Group cells
ad_cells = mic_metadata[mic_metadata['alzheimers_or_control'] == 1].index
ctrl_cells = mic_metadata[mic_metadata['alzheimers_or_control'] == 0].index

# Store results
deg_results = []

for gene in expressed_genes:
    ad_expr = mic_matrix.loc[ad_cells, gene]
    ctrl_expr = mic_matrix.loc[ctrl_cells, gene]
    
    if ad_expr.sum() == 0 and ctrl_expr.sum() == 0:
        continue
    #if ad_expr.sum() == 0 or ctrl_expr.sum() == 0:

    stat, p = mannwhitneyu(ad_expr, ctrl_expr, alternative='two-sided')
    effect = ad_expr.median() - ctrl_expr.median()
    if effect == 0:
        print(ad_expr, ctrl_expr)
    deg_results.append({'gene': gene, 'p_value': p, 'effect_size': effect})

deg_df = pd.DataFrame(deg_results).sort_values('p_value')
```

```python
deg_df.shape
```

```python
# FDR correction
deg_df['fdr'] = pd.Series(deg_df['p_value']).rank(method='min') / len(deg_df) * 0.05

# Filter for statistically significant genes
significant_degs = deg_df[deg_df['p_value'] < 1e-3]

# Print the top 10 by p-value
print("Top 10 DEGs by p-value between controls and AD:")
print(deg_df.head(10))
```

```python
mic_metadata = metadata[metadata['Subcluster'].isin(['Mic0', 'Mic1'])].copy()
mic_matrix = gene_matrix.loc[mic_metadata['TAG']]
mic_metadata = mic_metadata.set_index('TAG')
mic_data = mic_matrix.copy()
mic_data['Subcluster'] = mic_metadata['Subcluster']
```

```python
# Now check if these genes are differentially expressed between subclusters

from scipy.stats import mannwhitneyu

top_ad_genes = ['PLXDC2', 'SRGN', 'OLR1', 'SLC1A3', 'SDK1', 
                'RTTN', 'P2RY12', 'HS3ST4', 'SOCS6', 'KCNMA1']

results = []
for gene in top_ad_genes:
    mic0_expr = mic_data[mic_data['Subcluster'] == 'Mic0'][gene]
    mic1_expr = mic_data[mic_data['Subcluster'] == 'Mic1'][gene]

    # Skip genes with all zeros
    if mic0_expr.sum() == 0 and mic1_expr.sum() == 0:
        continue

    stat, pval = mannwhitneyu(mic0_expr, mic1_expr, alternative='two-sided')
    effect = mic1_expr.median() - mic0_expr.median()

    results.append({'gene': gene, 'pval': pval, 'effect_size': effect})

diff_subcluster_df = pd.DataFrame(results).sort_values('pval')
```

```python
print(diff_subcluster_df)
```

We performed targetted differential expression analysis between Mic0 and Mic1 subclusters and found that the above genes expression was signficantly different between them.

```python
import matplotlib.pyplot as plt
import numpy as np

# Use your existing dataframe
deg_df = df.copy()

# Calculate -log10(p-value)
deg_df['-log10(pval)'] = -np.log10(deg_df['pval'] + 1e-300)  # avoid log(0)

# Volcano plot
plt.figure(figsize=(8, 6))
plt.scatter(deg_df['log2fc'], deg_df['-log10(pval)'], 
            c='grey', s=10, alpha=0.7)

# Genes of intrest
genes_of_interest = ['PLXDC2', 'SLC1A3', 'SDK1', 'KCNMA1', 'CSGALNACT1', 'OLR1', 'P2RY12', 'HS3ST4', 'DOCK10', 'PICALM']
for gene in genes_of_interest:
    if gene in deg_df['gene'].values:
        row = deg_df[deg_df['gene'] == gene].iloc[0]
        plt.scatter(row['log2fc'], row['-log10(pval)'], color='red', s=40)
        plt.text(row['log2fc'], row['-log10(pval)'] + 0.1, gene, fontsize=9)

# Annotate thresholds
plt.axhline(-np.log10(0.05), color='black', linestyle='--', linewidth=1)
plt.axvline(1, color='black', linestyle='--', linewidth=1)
plt.axvline(-1, color='black', linestyle='--', linewidth=1)

# Labels
plt.xlabel('Log2 Fold Change')
plt.ylabel('-log10(p-value)')
plt.title('Volcano Plot (AD vs Control)')
plt.tight_layout()
plt.show()
```

```python
# Step 1: Subset to microglial cells
mic_df = adata.obs[adata.obs['Subcluster'].str.startswith('Mic')]

# Step 2: Count AD/control status per subcluster
subcluster_counts = mic_df.groupby(['Subcluster', 'alzheimers_or_control']).size().unstack(fill_value=0)

# Step 3: Calculate AD proportion per subcluster
subcluster_props = subcluster_counts.div(subcluster_counts.sum(axis=1), axis=0)
subcluster_props = subcluster_props.rename(columns={0: 'Control', 1: 'AD'})

# Step 4: Barplot for visual enrichment
subcluster_props.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])
plt.title('Proportion of AD vs Control in Microglial Subclusters')
plt.ylabel('Proportion')
plt.xticks(rotation=0)
plt.legend(title='Status')
plt.tight_layout()
plt.show()

# Step 5: Print enrichment ratio for Mic0 vs others
mic0_ad_ratio = subcluster_props.loc['Mic0', 'AD']
others_ad_ratio = subcluster_props.drop('Mic0')['AD'].mean()
print(f"Mic0 AD Proportion: {mic0_ad_ratio:.3f}")
print(f"Other Mics AD Proportion (average): {others_ad_ratio:.3f}")
```
