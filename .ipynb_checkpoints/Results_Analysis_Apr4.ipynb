Plotting Precision Recall Curve and ROC_AUC curve for the full model (keep for context that this model only assigned feature importance tot eh first 4 features)

```python
# Precision Recall Curve
from sklearn.metrics import precision_recall_curve, average_precision_score
import matplotlib.pyplot as plt

# Extract true labels and predicted probabilities
y_true = tgd_df['true_label']
y_scores = tgd_df['predicted_proba']

# Compute precision-recall curve
precision, recall, _ = precision_recall_curve(y_true, y_scores)
avg_precision = average_precision_score(y_true, y_scores)

# Plot
plt.plot(recall, precision, label=f'Avg Precision = {avg_precision:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# ROC Curve
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Get true labels and predicted probabilities
y_true = tgd_df['true_label']
y_scores = tgd_df['predicted_proba']

# Calculate ROC curve and AUC
fpr, tpr, _ = roc_curve(y_true, y_scores)
roc_auc = roc_auc_score(y_true, y_scores)

# Plot ROC
plt.plot(fpr, tpr, label=f'ROC AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Get row with 4 features
row_lower = tgd_inc[tgd_inc['num_features'] == 4].iloc[0]

# Get row with 17 features
row_upper = tgd_inc[tgd_inc['num_features'] == 15].iloc[0]

# Define metrics of interest
metrics = [
    'test_accuracy', 'test_roc_auc', 'test_avg_precision',
    'test_recall', 'test_precision', 'test_f1', 'test_mcc'
]

# Output for lower features
print(" Lower-Feature Classifier")
print("Features used:", row_lower['names_of_features'])
for metric in metrics:
    print(f"{metric}: {row_lower[metric]:.4f}")

print("\n Upper-Feature Classifier")
print("Features used:", row_upper['names_of_features'])
for metric in metrics:
    print(f"{metric}: {row_upper[metric]:.4f}")
```

Based on the above, it seems like jsut becuase the additonal features after the top 4 features hae a feature importance of zero does not mean that these features don't have some biological interaction. 

```python
print("\n Performance Gain from Adding Features:")
for metric in metrics:
    delta = row_upper[metric] - row_lower[metric]
    print(f"{metric}: {delta:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# List of metrics
metrics = [
    'test_accuracy', 'test_roc_auc', 'test_avg_precision',
    'test_recall', 'test_precision', 'test_f1', 'test_mcc'
]

# Extract values
lower_values = [row_lower[metric] for metric in metrics]
upper_values = [row_upper[metric] for metric in metrics]

# Set positions for bars
x = np.arange(len(metrics))
width = 0.35

# Plot
plt.figure(figsize=(10, 5))
plt.bar(x - width/2, lower_values, width, label='Lower-Feature Model (4 features)', color='skyblue')
plt.bar(x + width/2, upper_values, width, label='Upper-Feature Model (15 features)', color='orange')

# Labels & Formatting
plt.ylabel('Score')
plt.title('Comparison of Model Performance Metrics')
plt.xticks(x, metrics, rotation=45, ha='right')
plt.ylim(0, 1)
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load both prediction files
df_4 = pd.read_csv('/n/groups/patel/adithya/Alz_Outputs/Final_Outputs/Total_both/all_cell_types/test_predictions_top_4_features.csv')
df_4 = add_subcluster(df_4, metadata_file)

df_15 = pd.read_csv('/n/groups/patel/adithya/Alz_Outputs/Final_Outputs/Total_both/all_cell_types/test_predictions_top_15_features.csv')
df_15 = add_subcluster(df_15, metadata_file)

# Filter for AD cases (true_label == 1)
ad_4 = df_4[df_4['true_label'] == 1].copy()
ad_15 = df_15[df_15['true_label'] == 1].copy()

# Align both dataframes by TAG (cell identifier)
ad_4 = ad_4.set_index('TAG').sort_index()
ad_15 = ad_15.set_index('TAG').sort_index()

# Compute predicted probability difference
ad_15['delta_proba'] = ad_15['predicted_proba'] - ad_4['predicted_proba']

# Group by broad.cell.type and calculate average delta
avg_delta_by_celltype = ad_15.groupby('broad.cell.type')['delta_proba'].mean().sort_values(ascending=False)

# Plot
plt.figure(figsize=(10, 6))
ax = avg_delta_by_celltype.plot(kind='bar', color='steelblue')
plt.ylabel('Average Change in Predicted Probability (15 features - 4 features)')
plt.xlabel('Broad Cell Type')
plt.title('Effect of Additional Features on AD Prediction Confidence (for cases) by Cell Type')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

# Add text labels on top of bars
for i, value in enumerate(avg_delta_by_celltype):
    ax.text(i, value + 0.005, f'{value:.3f}', ha='center', va='bottom', fontsize=9)

plt.show()
```

This bar plot shows how much the models confidence in predicting Alzheimers (AD) increases per cell type when you go from using 4 demographic features to 15 features.  	Positive bars (e.g., Oli, Opc, Ast, Mic) mean that adding gene features makes the model more confident those cells are from AD patients. We must note that for any given patient, every cell had the same predicted probability, regardless of gene expression or cell identity becuase the feature importance of non-demographic features was treated as zero.


Although tree-based feature importance values assigned zero gain to gene expression features, adding these features changed the models predicted probabilities  particularly for specific cell types. For example, oligodendrocytes and microglia saw notable increases in AD prediction confidence. This suggests that, while the classifier may not have explicitly weighted these genes, their presence may have influenced model behavior.

```python
# Load gene matrix
gene_matrix = pd.read_parquet('/home/adm808/NormalizedCellMatrixSyn18485175.parquet').T

# Restrict to test microglial cells only (TAGs from Mic_full_df)
mic_tags = Mic_full_df['TAG'].values
mic_matrix = gene_matrix.loc[mic_tags]

# Use Mic_full_df directly as metadata
mic_metadata = Mic_full_df.set_index('TAG').loc[mic_tags]
```

```python
def select_missing_genes(matrix):
    mean_threshold = 2
    missingness_threshold = 90
    mean_expr = matrix.mean(axis=0)
    missingness = (matrix == 0).sum(axis=0) / matrix.shape[0] * 100
    low_genes = (missingness > missingness_threshold) & (mean_expr < mean_threshold)
    return matrix.columns[low_genes]

mic_matrix_filtered = mic_matrix.drop(columns=select_missing_genes(mic_matrix))
```

```python
import scanpy as sc

adata = sc.AnnData(mic_matrix_filtered)
adata.obs = mic_metadata.loc[adata.obs_names]
```

```python
ad_mic = adata[adata.obs['true_label'] == 1].copy()
```

```python
# Compute mean expression per subcluster
mean_spp1 = adata.obs.groupby('Subcluster')['SPP1_obs'].mean().sort_values(ascending=False)
mean_c1qa = adata.obs.groupby('Subcluster')['C1QA_obs'].mean().sort_values(ascending=False)

# Print the results
print("Mean SPP1 Expression per Subcluster:\n", mean_spp1.round(2), "\n")
print("Mean C1QA Expression per Subcluster:\n", mean_c1qa.round(2))
```

```python
from scipy.stats import ttest_ind

# Get values
mic1_spp1 = adata.obs.query("Subcluster == 'Mic1'")['SPP1_obs']
mic0_spp1 = adata.obs.query("Subcluster == 'Mic0'")['SPP1_obs']

mic1_c1qa = adata.obs.query("Subcluster == 'Mic1'")['C1QA_obs']
mic0_c1qa = adata.obs.query("Subcluster == 'Mic0'")['C1QA_obs']

# Run unpaired t-tests
spp1_p = ttest_ind(mic1_spp1, mic0_spp1, equal_var=False).pvalue
c1qa_p = ttest_ind(mic1_c1qa, mic0_c1qa, equal_var=False).pvalue

print(f"SPP1 (Mic1 vs Mic0) p-value: {spp1_p:.3e}")
print(f"C1QA (Mic1 vs Mic0) p-value: {c1qa_p:.3e}")
```
