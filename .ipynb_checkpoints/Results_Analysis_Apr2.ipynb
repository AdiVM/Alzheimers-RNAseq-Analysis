```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import joblib
```

```python
def misclass_results(predicitons, metadata, output):
    pred_total = pd.read_csv(predicitons)
    metadata_file = pd.read_parquet(metadata)
    df = add_subcluster(pred_total, metadata_file)

    mis_df = create_misclassified(pred_total, metadata_file)
    mis_df = add_subcluster(mis_df, metadata_file)
    mis_df.head()


    total_cells_per_cluster(df)
    total_cells_per_cluster(mis_df)


    misclassification_rate_per_cluster(df)

    total_metrics = pd.read_csv(output)

    print(total_metrics)
```

```python
pred_total = pd.read_csv("/n/groups/patel/adithya/Alz_Outputs/Final_Outputs/Total_both/all_cell_types/train_predictions.csv")
metadata_file = pd.read_parquet('/home/adm808/New_CellMetadataSyn1848517.parquet')
df = add_subcluster(pred_total, metadata_file)

mis_df = create_misclassified(pred_total, metadata_file)
mis_df = add_subcluster(mis_df, metadata_file)
mis_df.head()


total_cells_per_cluster(df)
total_cells_per_cluster(mis_df)


misclassification_rate_per_cluster(df)

total_metrics = pd.read_csv("/n/groups/patel/adithya/Alz_Outputs/Final_Outputs/Total_both/all_cell_types/output_csv.csv")

total_metrics
```
